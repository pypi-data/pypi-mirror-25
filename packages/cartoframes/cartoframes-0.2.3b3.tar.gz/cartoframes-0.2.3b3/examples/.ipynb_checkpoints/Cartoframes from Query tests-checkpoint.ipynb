```python
import pandas as pd
import cartoframes
import json

USERNAME = json.loads(open('credentials.json').read())['username']
APIKEY = json.loads(open('credentials.json').read())['api_key']
```

### df from query without table (no linkage)

```python
df_from_query = pd.read_carto(username=USERNAME,
                              api_key=APIKEY,
                              query='SELECT * FROM all_month_3 LIMIT 10')
df_from_query.head(2)
```

### metadata should be empty

```python
df_from_query._metadata
```

### as a result, maps should error with message about how to link with carto

```python
df_from_query.carto_map(interactive=True)
```

### link to carto with new tablename

```python
df_from_query.sync_carto(username=USERNAME,
                         api_key=APIKEY,
                         requested_tablename='cartoframe_from_query_test_march_13',
                         is_org_user=True,
                         debug=True)
```
